**DESCRIPTIVE STATASTICS**

```python
mean=df.mean(numeric_only=True)
median=df.median(numeric_only=True)
mode=df.mode().iloc[0]
std=df.std(numeric_only=True)
print("mean values:\n",mean)
print("median values:\n",median)
print("mode values:\n",mode)
print("standard deviation values:\n",std)
```

TIME SERIES ANALYSIS


```python
sales_trends=df.groupby('Date')['Total Amount'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales_trends.index,sales_trends.values,marker='o',linestyle="-")
plt.title("sales trends over time")
plt.xlabel('Date')
plt.ylabel('Toatal Sales Amount')
plt.tight_layout()
plt.show()
```

CUSTOMER DEMOGRAPHICS


```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['Age'],bins=15,kde=True)
plt.title('age distribution')
plt.subplot(1,2,2)
sns.countplot(x='Gender',data=df)
plt.title('Gender distribution')
```

Heat maps of sales by age and gender

```python
tot=df.pivot_table(index='Age',columns='Gender',values='Total Amount',aggfunc='sum')
plt.figure(figsize=(10,8))
sns.heatmap(tot,cmap='coolwarm',annot=True,fmt='.0f')
plt.title('heat map of sales by age and gender')
plt.show()
```
