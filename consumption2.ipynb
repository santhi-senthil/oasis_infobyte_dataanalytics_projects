```python
def get_relationship(r):
  if r['marital_Married']==1:
    return 1
  elif r['marital_Together']==1:
     return 1
  else:
    return 0
df['In_relationship']=df.apply(get_relationship,axis=1)
df.head()
```

```python
def edu_sts(r):
  if r['education_2n Cycle']==1:
    return 2
  elif r['education_Basic']==1:
    return 1
  elif r['education_Graduation']==1:
    return 3
  elif r['education_Master']==1:
    return 4
  elif r['education_PhD']==1:
    return 5
  else:
    return 'unknown'
df['education_sts']=df.apply(edu_sts,axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
cols_cluster=['Income','MntTotal','In_relationship','education_sts']
df_scaled=df.copy()
df_scaled[cols_cluster]= scaler.fit_transform(df[cols_cluster])
```

```python
x=df_scaled[cols_cluster]
inertia_list=[]
for k in range(1,11):
  inertia=KMeans(n_clusters=k ,n_init=10).fit(x).inertia_
  inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(1,11),inertia_list)
plt.title("Inertia vs number cluster")
plt.xlabel("numbers of clusters (k)")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn import decomposition
pca=decomposition.PCA(n_components=2)
pca_res=pca.fit_transform(df_scaled[cols_cluster])
df_scaled['pc1']=pca_res[:,0]
df_scaled['pc2']=pca_res[:,1]
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='pc1',y='pc2',data=df_scaled,palette='viridis')
plt.title('cluster_visuaization')
plt.xlabel("principal component 1(pc1)")
plt.ylabel("principal component 2(pc2)")
plt.legend(title='clusters')
```

```python
model=KMeans(n_clusters=4,random_state=7)
model.fit(df_scaled[cols_cluster])
df_scaled['cluster']=model.predict(df_scaled[cols_cluster])
```

```python
cus_num=df_scaled.groupby('cluster')['MntTotal'].count().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='cluster',y='MntTotal',data=cus_num,palette='viridis')
plt.title('clustersizes')
plt.xlabel('cluster')
plt.ylabel('cus_num')
plt.show()
```

```python
scaler=StandardScaler()
cols_cluster=['Income','MntTotal','In_relationship','education_sts']
df_clusterd=df.copy()

model=KMeans(n_clusters=4,random_state=7)
model.fit(df_clusterd[cols_cluster])
df_clusterd['cluster']=model.predict(df_clusterd[cols_cluster])
df_clusterd.head()
```

```python
df_clusterd['avg_income']=df_clusterd.groupby('cluster')['Income'].mean()
plt.figure(figsize=(8,6))
sns.barplot(x='cluster',y='avg_income',data=df_clusterd,palette='viridis')
plt.title("cluster average income")
plt.xlabel('clusters')
plt.ylabel('avg_income')
plt.show()
```
